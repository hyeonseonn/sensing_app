## 1. CNN_LSTM (20Í∞ú Í≥ºÍ±∞ ‚Üí 5Í∞ú ÎØ∏Îûò)

```python
import numpy as np
from scipy.signal import sawtooth
import matplotlib.pyplot as plt

np.random.seed(42)
days = 90
points_per_day = 24 * 12
total_points = days * points_per_day

# ÏãúÍ∞ÑÏ∂ï
time = np.linspace(0, 1, total_points)

# üåÄ mix_waveÎ•º ÎÇ†ÏßúÎ≥ÑÎ°ú Îã§Î•∏ ÌååÌòïÏúºÎ°ú ÏÉùÏÑ± (Î∂àÍ∑úÏπôÏÑ± Í∞ïÌôî)
mix_wave = np.zeros(total_points)
for d in range(days):
    offset = np.random.uniform(0, 2*np.pi)
    amp = np.random.uniform(0.8, 1.2)
    freq = np.random.uniform(0.8, 1.3)
    start = d * points_per_day
    end = (d+1) * points_per_day
    t = np.linspace(0, 2*np.pi, points_per_day)
    wave = (
        amp * np.sin(freq * t + offset) +
        0.3 * sawtooth(freq * t + offset) +
        0.2 * (2 * np.abs((t / np.pi) % 2 - 1) - 1)
    )
    wave += np.random.normal(0, 0.1, points_per_day)
    mix_wave[start:end] = wave

# üéØ trendÎèÑ ÎÇ†ÎßàÎã§ Îã§Î•¥Í≤å ÏÉùÏÑ±
trend = np.zeros(total_points)
for d in range(days):
    start = d * points_per_day
    end = (d + 1) * points_per_day
    slope = np.random.uniform(-5, 5)
    curve = np.random.uniform(0, 3)
    t = np.linspace(0, 1, points_per_day)
    trend[start:end] = slope * t + curve * (t - 0.5)**2

# ÏóÖÎ¨¥ÏãúÍ∞Ñ ÎßàÏä§ÌÅ¨
work_hours = np.zeros(total_points)
for d in range(days):
    start_idx = d * points_per_day + 8 * 12
    end_idx = d * points_per_day + 18 * 12
    work_hours[start_idx:end_idx] = 1.0

# üìà ÏóÖÎ¨¥ÏãúÍ∞Ñ Î∂ÄÏä§ÌåÖÎèÑ ÎÇ†ÏßúÎ≥ÑÎ°ú ÎûúÎç§Ìïú Î≤îÏúÑÏóêÏÑú Îã§Î•¥Í≤å Ï†ÅÏö©
boost_add = np.zeros(total_points)
for d in range(days):
    boost_level = np.random.uniform(20, 30)
    start_idx = d * points_per_day + 8 * 12
    end_idx = d * points_per_day + 18 * 12
    boost_add[start_idx:end_idx] = boost_level + np.random.normal(0, 5, end_idx - start_idx)

# ‚úÖ ÏµúÏ¢Ö CPU ÏÇ¨Ïö©Î•†: Îçî Î∂àÍ∑úÏπôÌïòÍ≤å Íµ¨ÏÑ±
cpu = 45 + 10 * mix_wave + trend + boost_add

# ‚úÖ ÏóÖÎ¨¥ÏãúÍ∞Ñ ÌïòÌïúÏÑ† Î≥¥Ï†ï
cpu = np.where((work_hours == 1) & (cpu < 75), np.random.uniform(75, 90, size=total_points), cpu)

# ‚úÖ Ï†ÑÏ≤¥ ÌïòÌïúÏÑ† Î≥¥Ï†ï (ÏûêÏó∞Ïä§Îü¨ÏõÄ Ïú†ÏßÄ)
cpu = np.where(cpu < 30, 30 + (30 - cpu) * 0.3, cpu)

# ‚úÖ ÌèâÍ∑† Í≥ÑÏÇ∞
work_cpu = cpu[work_hours == 1.0]
nowork_cpu = cpu[work_hours == 0.0]
avg_work_cpu = np.mean(work_cpu)
avg_nowork_cpu = np.mean(nowork_cpu)
avg_cpu = max(avg_work_cpu, avg_nowork_cpu)
condition1 = avg_cpu >= 80

# ‚úÖ Í≤∞Í≥º Ï∂úÎ†•
print(f"üìä ÏóÖÎ¨¥ÏãúÍ∞Ñ ÌèâÍ∑† CPU ÏÇ¨Ïö©Î•†: {avg_work_cpu:.2f}%")
print(f"üìä ÎπÑÏóÖÎ¨¥ÏãúÍ∞Ñ ÌèâÍ∑† CPU ÏÇ¨Ïö©Î•†: {avg_nowork_cpu:.2f}%")
print(f"‚úÖ Ï°∞Í±¥ 1 ÎßåÏ°± Ïó¨Î∂Ä (Îëò Ï§ë Îçî ÌÅ∞ Í∞í Í∏∞Ï§Ä ‚â• 80%): {condition1} ({avg_cpu:.2f}%)")

# ‚úÖ ÏãúÍ∞ÅÌôî (5ÏùºÎßå)
plt.figure(figsize=(14, 4))
plt.plot(cpu)
plt.title("üìà ÏûêÏó∞Ïä§Îü¨Ïö¥ CPU ÏÇ¨Ïö©Î•† (Ï°∞Í±¥ 1 ÎßåÏ°±)")
plt.xlabel("ÏãúÍ∞Ñ (5Ïùº)")
plt.ylabel("CPU ÏÇ¨Ïö©Î•† (%)")
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)
days = 90
points_per_day = 24 * 12
total_points = days * points_per_day

# ÏóÖÎ¨¥ÏãúÍ∞Ñ ÎßàÏä§ÌÅ¨
work_hours = np.zeros(total_points)
for d in range(days):
    start = d * points_per_day + 8 * 12
    end = d * points_per_day + 18 * 12
    work_hours[start:end] = 1.0

# üîª mix_wave Ï†úÍ±∞ ÏàòÏ§ÄÏúºÎ°ú Ï∂ïÏÜå
mix_wave = np.random.normal(0, 0.001, total_points)  # Í±∞Ïùò ÌèâÌèâÌïú ÌååÌòï

# üîª trend ÏôÑÏ†Ñ Ï†úÍ±∞
trend = np.zeros(total_points)

# üîª boost: ÌèâÍ∑†Îßå Ïú†ÏßÄ, Î∂ÑÏÇ∞ÏùÄ Í∑πÏÜåÌôî
boost_add = np.zeros(total_points)
for d in range(days):
    # ÏóÖÎ¨¥ÏãúÍ∞Ñ boost: Í≥†Ï†ïÍ∞í + ÏïÑÏ£º ÏûëÏùÄ ÎÖ∏Ïù¥Ï¶à
    work_boost = np.random.uniform(24, 25)
    start = d * points_per_day + 8 * 12
    end = d * points_per_day + 18 * 12
    boost_add[start:end] = work_boost + np.random.normal(0, 0.05, end - start)

    # ÎπÑÏóÖÎ¨¥ÏãúÍ∞Ñ boost: Í≥†Ï†ïÍ∞í + ÏïÑÏ£º ÏûëÏùÄ ÎÖ∏Ïù¥Ï¶à
    nowork_boost = np.random.uniform(15, 20)
    start1 = d * points_per_day
    end1 = d * points_per_day + 8 * 12
    boost_add[start1:end1] = nowork_boost + np.random.normal(0, 0.05, end1 - start1)

    start2 = d * points_per_day + 18 * 12
    end2 = (d+1) * points_per_day
    boost_add[start2:end2] = nowork_boost + np.random.normal(0, 0.05, end2 - start2)

# ‚úÖ ÏµúÏ¢Ö CPU
cpu = 5 + mix_wave + trend + boost_add  # Í±∞Ïùò flat

# ‚úÖ ÌïòÌïúÏÑ† Î≥¥Ï†ï
cpu = np.where(cpu < 5, 5, cpu)

# ‚úÖ ÌèâÍ∞Ä ÏßÄÌëú
work_cpu = cpu[work_hours == 1.0]
nowork_cpu = cpu[work_hours == 0.0]
avg_work_cpu = np.mean(work_cpu)
avg_nowork_cpu = np.mean(nowork_cpu)
std_total = np.std(cpu)

condition1 = avg_work_cpu <= 30
condition2 = avg_work_cpu > avg_nowork_cpu
condition3 = std_total <= 5
all_conditions = condition1 and condition2 and condition3

# ‚úÖ Í≤∞Í≥º Ï∂úÎ†•
print(f"üìä ÏóÖÎ¨¥ÏãúÍ∞Ñ ÌèâÍ∑† CPU ÏÇ¨Ïö©Î•†: {avg_work_cpu:.2f}%")
print(f"üìä ÎπÑÏóÖÎ¨¥ÏãúÍ∞Ñ ÌèâÍ∑† CPU ÏÇ¨Ïö©Î•†: {avg_nowork_cpu:.2f}%")
print(f"üìä Ï†ÑÏ≤¥ ÌëúÏ§ÄÌé∏Ï∞®: {std_total:.4f}")
print(f"‚úÖ Ï∂ïÏÜå Ï°∞Í±¥ ÎßåÏ°± Ïó¨Î∂Ä: {all_conditions}")

# ‚úÖ ÏãúÍ∞ÅÌôî
plt.figure(figsize=(14, 4))
plt.plot(cpu, label='CPU Usage (%)', linewidth=0.7)
plt.axhline(avg_work_cpu, color='blue', linestyle='--', label='Work Avg')
plt.axhline(avg_nowork_cpu, color='red', linestyle='--', label='Nowork Avg')
plt.title("üìâ Ï¥àÏ†ÄÎ≥ÄÎèô CÎì±Í∏â CPU ÏãúÎÇòÎ¶¨Ïò§ (ÌëúÏ§ÄÌé∏Ï∞® ‚â§ 5%)")
plt.xlabel("ÏãúÍ∞Ñ (90Ïùº)")
plt.ylabel("CPU ÏÇ¨Ïö©Î•† (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# === Realistic Memory Leak: nightly upward trend with noise + occasional dips ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# 0) ÏãúÎÆ¨Î†àÏù¥ÏÖò Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
def simulate_realistic_memory(
    days=6,
    start_value=52.0,
    freq="5min",
    leak_per_hour=0.12,         # ÎπÑÏóÖÎ¨¥ÏãúÍ∞Ñ ÎàÑÏàò ÏÜçÎèÑ(+%/h)
    work_mu_per_hour=-0.02,     # ÏóÖÎ¨¥ÏãúÍ∞Ñ ÌèâÍ∑† Î≥ÄÌôî(-%/h: ÏÇ¥Ïßù ÎÇ¥Î†§Í∞ÄÍ≤å)
    gc_prob=0.015,              # ÎπÑÏóÖÎ¨¥ÏãúÍ∞Ñ ÏÜåÍ∑úÎ™® GC ÌïòÍ∞ï ÌôïÎ•†
    gc_drop_range=(0.05, 0.20), # GC ÌïòÍ∞ïÌè≠(%)
    seed=123
) -> pd.DataFrame:
    rng = np.random.default_rng(seed)
    idx = pd.date_range("2025-07-01", periods=days*24*12, freq=freq)
    mem = np.zeros(len(idx), dtype=float)
    mem[0] = start_value

    leak_step_mu  = leak_per_hour / (60 / pd.Timedelta(freq).seconds * 60)  # per step
    leak_step_sd  = leak_step_mu * 0.6
    work_step_mu  = work_mu_per_hour / (60 / pd.Timedelta(freq).seconds * 60)
    work_step_sd  = abs(work_step_mu) * 1.5 + 0.02  # ÏóÖÎ¨¥ÏãúÍ∞Ñ ÎÖ∏Ïù¥Ï¶à

    for i in range(1, len(idx)):
        h = idx[i].hour
        if (h >= 18 or h < 8):  # ÎπÑÏóÖÎ¨¥ÏãúÍ∞Ñ
            inc = rng.normal(leak_step_mu, leak_step_sd)
            if rng.random() < gc_prob:
                inc -= rng.uniform(*gc_drop_range)  # ÏûëÏùÄ ÌïòÍ∞ï(Í∑∏ÎûòÎèÑ ÏàúÏ¶ùÍ∞ÄÍ∞Ä Ïö∞ÏÑ∏)
        else:  # ÏóÖÎ¨¥ÏãúÍ∞Ñ
            inc = rng.normal(work_step_mu, work_step_sd)
        mem[i] = max(0.0, mem[i-1] + inc)

    return pd.DataFrame({"memory": mem}, index=idx)

# 1) ÎàÑÏàò Í∞êÏßÄ: ÎπÑÏóÖÎ¨¥ÏãúÍ∞Ñ(18:00~08:00) Íµ¨Í∞Ñ ÌöåÍ∑Ä Í∏∞Ïö∏Í∏∞ > 0
def detect_memory_leak(memory_df: pd.DataFrame,
                       work_start="08:00",
                       work_end="18:00",
                       slope_threshold=0.0,
                       min_points=5):
    df = memory_df.copy()
    if not isinstance(df.index, pd.DatetimeIndex):
        if "timestamp" in df.columns:
            df["timestamp"] = pd.to_datetime(df["timestamp"])
            df = df.set_index("timestamp")
        else:
            raise ValueError("DatetimeIndex ÎòêÎäî 'timestamp' Ïª¨ÎüºÏù¥ ÌïÑÏöîÌï©ÎãàÎã§.")
    df = df.sort_index()

    unique_days = pd.Index(sorted(pd.to_datetime(df.index.date).unique()))
    rows, leak_days = [], []

    for d in unique_days[:-1]:  # ÎßàÏßÄÎßâ ÎÇ†ÏùÄ Îã§ÏùåÎÇ† 08:00Ïù¥ ÏóÜÏùÑ Ïàò ÏûàÏùå
        start = pd.Timestamp(d) + pd.Timedelta(hours=int(work_end.split(":")[0]))
        end   = pd.Timestamp(d) + pd.Timedelta(days=1) + pd.Timedelta(hours=int(work_start.split(":")[0]))

        seg = df.loc[(df.index >= start) & (df.index < end), "memory"]
        if len(seg) < min_points:
            continue

        x = (seg.index - seg.index[0]).total_seconds() / 3600.0  # ÏãúÍ∞Ñ Îã®ÏúÑ
        y = seg.values.astype(float)
        slope, intercept, r, p, stderr = linregress(x, y)
        is_leak = slope > slope_threshold
        rows.append({"date": d, "start": start, "end": end,
                     "slope": slope, "r": r, "p": p, "is_leak": is_leak})
        if is_leak:
            leak_days.append(d)

    detail = pd.DataFrame(rows)
    return {"is_memory_leak": len(leak_days) > 0, "leak_days": leak_days, "detail": detail}

# 2) Ï†ÑÏ≤¥ ÏãúÍ≥ÑÏó¥ + ÎàÑÏàò ÏïºÍ∞Ñ ÌöåÍ∑ÄÏÑ† Ïò§Î≤ÑÎ†àÏù¥
def plot_memory_with_leak(memory_df: pd.DataFrame, detail_df: pd.DataFrame):
    df = memory_df.copy()
    if not isinstance(df.index, pd.DatetimeIndex):
        if "timestamp" in df.columns:
            df["timestamp"] = pd.to_datetime(df["timestamp"])
            df = df.set_index("timestamp")
        else:
            raise ValueError("DatetimeIndex ÎòêÎäî 'timestamp' Ïª¨ÎüºÏù¥ ÌïÑÏöîÌï©ÎãàÎã§.")
    df = df.sort_index()

    plt.figure(figsize=(12, 4))
    plt.plot(df.index, df["memory"], label="Memory Usage (%)")

    for _, row in detail_df[detail_df["is_leak"]==True].iterrows():
        seg = df.loc[(df.index >= row["start"]) & (df.index < row["end"]), "memory"]
        if len(seg) < 2:
            continue
        x = (seg.index - seg.index[0]).total_seconds()/3600.0
        y_fit = row["slope"]*x + seg.values[0]
        plt.plot(seg.index, y_fit, linewidth=2,
                 label=f"Leak {row['date'].date()} (slope={row['slope']:.3f}/h)")

    # plt.title("Realistic Memory Leak (non-working hours trend up)")
    plt.xlabel("Time"); plt.ylabel("Memory Usage (%)"); plt.legend(); plt.tight_layout(); plt.show()

# 3) (ÏòµÏÖò) ÌäπÏ†ï Î∞§ Íµ¨Í∞ÑÎßå ÌôïÎåÄ
def plot_zoom_one_night(memory_df: pd.DataFrame, night_row: pd.Series):
    seg = memory_df.loc[(memory_df.index >= night_row["start"]) & (memory_df.index < night_row["end"]), "memory"]
    x = (seg.index - seg.index[0]).total_seconds()/3600.0
    y_fit = night_row["slope"]*x + seg.values[0]

    plt.figure(figsize=(12, 4))
    plt.plot(seg.index, seg.values, label=f"Night window {night_row['date'].date()}")
    plt.plot(seg.index, y_fit, linewidth=2, label=f"Regression slope={night_row['slope']:.3f}/h")
    plt.title("Zoom: Non-working hours (noisy, occasional dips, but net upward)")
    plt.xlabel("Time"); plt.ylabel("Memory (%)"); plt.legend(); plt.tight_layout(); plt.show()

# ---------------------- ÏÇ¨Ïö© ÏòàÏãú ----------------------
# (A) ÏÉòÌîå Îç∞Ïù¥ÌÑ∞Î°ú ÌÖåÏä§Ìä∏
memory_df = simulate_realistic_memory(
    days=6,
    leak_per_hour=0.12,      # ‚Üë ÏïºÍ∞Ñ ÎàÑÏàò ÏÜçÎèÑ
    work_mu_per_hour=-0.02,  # ÎÇÆÏóêÎäî ÏÇ¥Ïßù ÎÇ¥Î†§Í∞ÄÍ±∞ÎÇò Ïú†ÏßÄ
    gc_prob=0.015,
    gc_drop_range=(0.05, 0.20)
)

# (B) ÎÇ¥ Îç∞Ïù¥ÌÑ∞ ÏÇ¨Ïö© Ïãú
# memory_df = pd.read_csv("memory.csv", parse_dates=["timestamp"]).set_index("timestamp")
# memory_df.rename(columns={"<Î©îÎ™®Î¶¨Ïª¨ÎüºÎ™Ö>": "memory"}, inplace=True)

# ÎàÑÏàò Í∞êÏßÄ & ÏãúÍ∞ÅÌôî
res = detect_memory_leak(memory_df, work_start="08:00", work_end="18:00", slope_threshold=0.0)
print("Î©îÎ™®Î¶¨ ÎàÑÏàò ÌåêÎã®:", res["is_memory_leak"], "ÎàÑÏàò ÏùºÏàò:", len(res["leak_days"]))
print(res["detail"].head())

plot_memory_with_leak(memory_df, res["detail"])

# ÌôïÎåÄ Í∑∏ÎûòÌîÑ(ÏòµÏÖò)
if not res["detail"].empty:
    night = res["detail"].iloc[0]
    plot_zoom_one_night(memory_df, night)
```
